```python
import matplotlib.pyplot as plt
import numpy as np

import palettetown.pokeplot as pkp
```

```python
def make_smooth_data():
    l = np.linspace(-2,2,100)
    xs, ys = np.meshgrid(l,l)
    rs = np.sqrt(xs**2 + ys**2)
    return np.exp(-rs)

smooth_data = make_smooth_data()
random_data = np.random.random((10,10))

pokemon_names = ('zubat','linoone','metang','groudon','xerneas')
_,axs = plt.subplots(3,len(pokemon_names),figsize=(3*len(pokemon_names),10))
for pokemon_name,axcol in zip(pokemon_names,axs.T):
    # draw a pokemon sprite
    pkp.draw_sprite(pokemon_name,ax=axcol[0])
    # get a colormap for the pokemon
    cmap = pkp.get_cmap(pokemon_name)
    # plot using the colormap
    axcol[1].imshow(smooth_data,cmap=cmap)
    axcol[2].imshow(random_data,cmap=cmap)
```
